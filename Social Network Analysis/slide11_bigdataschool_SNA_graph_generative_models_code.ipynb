```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

#### Read zachary dataset graph

```python
#zachary = nx.read_edgelist('zachary.csv', delimiter=',', nodetype=int)
zachary = nx.read_edgelist('zachary.txt',  nodetype=int)
# Print number of nodes and edges
print(zachary)
print(zachary.nodes())
```

#### Plot degree distribution

```python
# Get the degree of each node
degrees = [zachary.degree(node) for node in zachary.nodes()]
print(degrees)
```

```python
# Plot the degree distribution with larger figure size
fig, ax = plt.subplots(figsize=(6, 4))

ax.hist(degrees, bins=range(max(degrees)+2), align='mid', rwidth=0.8)
ax.set_xlabel("Degree")
ax.set_ylabel("Frequency")
ax.set_title("Degree distribution of the Zachary karate club graph")
plt.show()
```

#### Plot log-log degree distribution

```python
# Calculate the degree frequency
degree_freq = np.bincount(degrees)
print(degree_freq)
```

```python

# Create a log-log scatter plot of the degree distribution
fig, ax = plt.subplots(figsize=(6, 4))
plt.scatter(range(len(degree_freq)), degree_freq, marker='o')
plt.xscale('log')
plt.yscale('log')

# Set the plot attributes
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

#### Calculate clustering coefficient for all graph nodes

```python
# Calculate the overall clustering coefficient for the graph
occ = nx.average_clustering(zachary)

# Print the the overall clustering coefficient for the graph
print("Overall clustering coefficient for the graph: {:0.4f}\n".format(occ))
```

```python
# Calculate the clustering coefficient for each node
clustering_coefficients = nx.clustering(zachary)

# Print the clustering coefficient for each node
print("Clustering coefficient for all nodes")
for node, cc in clustering_coefficients.items():
    print("Node {}: {:0.4f}".format(node, cc), end="   ,   ")
```
