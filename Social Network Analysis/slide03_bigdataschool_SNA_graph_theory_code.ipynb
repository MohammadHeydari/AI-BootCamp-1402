```python
import networkx as nx
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Create First Graph

```python
# Create a new graph object
graph01 = nx.Graph()

# Add 8 nodes to the graph
graph01.add_nodes_from(range(1, 9))

# Add 12 random edges to the graph
nodes = list(graph01.nodes())
while graph01.number_of_edges() < 12:
    u, v = random.sample(nodes, 2)
    if not graph01.has_edge(u, v):
        graph01.add_edge(u, v)

print(graph01)
```

#### Visualize the graph

```python
# Create the subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# First layout: spring layout
pos1 = nx.spring_layout(graph01)
nx.draw(graph01, pos1, with_labels=True,
        font_color="white", font_size=10,
        node_color="darkblue", edgecolors="black",
        font_weight="bold", ax=axs[0])
# nx.draw_networkx_edge_labels(graph01, pos1, edge_labels={}, ax=axs[0])
axs[0].set_title('Spring Layout')

# Second layout: circular layout
pos2 = nx.circular_layout(graph01)
nx.draw(graph01, pos2, with_labels=True,
        font_color="white", font_size=10,
        node_color="darkblue", edgecolors="black",
        font_weight="bold", ax=axs[1])
# nx.draw_networkx_edge_labels(graph01, pos2, edge_labels={}, ax=axs[1])
axs[1].set_title('Circular Layout')

# Third layout: random layout
pos3 = nx.random_layout(graph01)
nx.draw(graph01, pos3, with_labels=True,
        font_color="white", font_size=10,
        node_color="darkblue", edgecolors="black",
        font_weight="bold", ax=axs[2])
# nx.draw_networkx_edge_labels(graph01, pos3, edge_labels={}, ax=axs[2])
axs[2].set_title('Random Layout')

plt.show()
```

#### Show edge list of the graph

```python
print("Edge list of graph01:")
for edge in graph01.edges():
    print(edge)
```

#### Show adjacency list of graph

```python
print("Adjacency list of graph01:")
for node in graph01.nodes():
    neighbors = ", ".join(str(n) for n in graph01.neighbors(node))
    print(f"{node}: {neighbors}")
```

#### show adjacency matrix of graph

```python
# Create an adjacency matrix
adj_matrix = nx.adjacency_matrix(graph01).todense()

# Convert the matrix to a pandas DataFrame
adj_matrix_df = pd.DataFrame(adj_matrix)

# Print the adjacency matrix
print('Adjacency Matrix = \n')
print(adj_matrix_df)
```

#### Print components and jiant component of the graph

```python
# Read the graph in GraphML format from a file
graph02 = nx.read_graphml("graph02.graphml")

# Print graph nodes
print(graph02.nodes())

# First layout: spring layout
pos1 = nx.spring_layout(graph02)
nx.draw(graph02, pos1, with_labels=True,
        font_color="white", font_size=10,
        node_color="darkblue", edgecolors="black",
        font_weight="bold")
plt.show()

```

```python
# Convert node IDs from strings to integers
graph02 = nx.relabel.convert_node_labels_to_integers(graph02, first_label=0)

# Print graph nodes
print(graph02.nodes())
```

```python
# Find all connected components of the graph
components = list(nx.connected_components(graph02))

# Print the nodes in each connected component
i = 1
for component in components:
    print(f"Nodes in component {i} has {len(component)} nodes: {list(component)}")
    i += 1
```

```python
# Find the size of the largest component
largest_component = max(components, key=len)

# Create a subgraph containing only the largest component
giant_component = graph02.subgraph(largest_component)

# Print the nodes and edges in the giant component
print(f"Nodes in the giant component with {len(giant_component.nodes())} nodes: {giant_component.nodes()}")
print(f"Edges in the giant component with {len(giant_component.edges())} edges: {giant_component.edges()}")
```
