```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Create and visualize bipartite graph

```python
# Create a bipartite graph
graph04 = nx.Graph()

# Define first partition
first_partition_nodes = ['Mehran', 'Ali', 'Pouya', 'Mohammad', 'Sina', 'Amir']

# Define second partition
second_partition_nodes = ['Apple', 'Banana', 'Cherry', 'Kiwi', 'Lemon', 'Mango', 
                          'Peach', 'Pineapple', 'Watermelon']

# Add nodes from first partition
graph04.add_nodes_from(first_partition_nodes, bipartite=0)

# Add nodes from second partition
graph04.add_nodes_from(second_partition_nodes, bipartite=1)

# Add edges between two partition
graph04.add_edges_from([('Mehran', 'Mango'), ('Mehran', 'Pineapple'), 
                        ('Ali', 'Apple'), ('Ali', 'Banana'), ('Ali', 'Kiwi'), ('Ali', 'Mango'), ('Ali', 'Watermelon'), ('Ali', 'Peach'), ('Ali', 'Lemon'),
                        ('Pouya', 'Apple'), ('Pouya', 'Cherry'), ('Pouya', 'Lemon'), ('Pouya', 'Watermelon'),
                        ('Mohammad', 'Banana'), ('Mohammad', 'Lemon'), ('Mohammad', 'Apple'),
                        ('Sina', 'Peach'), ('Sina', 'Pineapple'), ('Sina', 'Kiwi'), 
                        ('Amir', 'Cherry'), ('Amir', 'Lemon'),
                        ])

# Print number of nodes and edges
print(graph04)
```

#### visualize the bipartite graph

```python
plt.figure(figsize=(10, 10))
pos = nx.bipartite_layout(graph04, nodes=first_partition_nodes)  


# draw nodes  
nx.draw_networkx_nodes(graph04, pos, nodelist=first_partition_nodes, node_color='r', node_size=500)
nx.draw_networkx_nodes(graph04, pos, nodelist=second_partition_nodes, node_color='b', node_size=500)

# draw edges
nx.draw_networkx_edges(graph04, pos)  

# add node labels
label_pos = {k:(v[0], v[1]+0.05) for k,v in pos.items()}
nx.draw_networkx_labels(graph04, label_pos, font_size=12, font_weight='bold', font_color='k')
plt.axis("off")
plt.title("a bipartite graph")
plt.show()
```

#### Convert bipartite graph to one-partite graph

```python
# Convert the bipartite graph to a one-partite graph
one_partite_graph = nx.bipartite.projected_graph(graph04, first_partition_nodes)
print(one_partite_graph)
```

#### visualize the one-partite graph

```python
# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# plot the one-partite graph in the first subplot
pos1 = nx.kamada_kawai_layout(one_partite_graph)

# remove the border around the first subplot
ax1.set_frame_on(False)

# Set the nodes label position
label_pos = {k:(v[0]+0.02, v[1]+0.06) for k,v in pos1.items()}

# add a title to the first subplot
ax1.set_title('One-Partite Graph', fontsize=16)

# Set the label for nodes
nx.draw_networkx_labels(one_partite_graph, label_pos,
                        font_size=12, font_weight='bold',
                        font_color='k', ax=ax1)

# Draw the graph
nx.draw_networkx(one_partite_graph, pos=pos1,
                 font_weight='bold', with_labels=False,
                 node_color="red", node_size=500,
                 edgecolors="black", ax=ax1)

# plot the second graph in the second subplot
pos2 = nx.bipartite_layout(graph04, nodes=first_partition_nodes)

# remove the border around the first subplot
ax2.set_frame_on(False)

# draw nodes  
nx.draw_networkx_nodes(graph04, pos2, nodelist=first_partition_nodes, node_color='r', node_size=500, ax=ax2)
nx.draw_networkx_nodes(graph04, pos2, nodelist=second_partition_nodes, node_color='b', node_size=500, ax=ax2)

# draw edges
nx.draw_networkx_edges(graph04, pos2, ax=ax2)

# add node labels
label_pos = {k:(v[0], v[1]+0.053) for k,v in pos.items()}
nx.draw_networkx_labels(graph04, label_pos, font_size=12, font_weight='bold', font_color='k', ax=ax2)

# add a title to the first subplot
ax2.set_title('Bipartite Graph', fontsize=16)

plt.show()
```
