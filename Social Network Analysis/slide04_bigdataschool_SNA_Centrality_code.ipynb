```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Load zachary karate club from edge list

```python
# Create a graph from the edge list file using read_edgelist()
zachary = nx.read_edgelist("zachary.txt")

# Convert node IDs from string to integer
zachary = nx.relabel.convert_node_labels_to_integers(zachary, first_label=0)

# Print number of nodes and edges
print(zachary)
```

#### Visualize zachary karate club

```python
# Set the layout for graph plots
pos = nx.kamada_kawai_layout(zachary)

# plot the graph
plt.figure(figsize=(12, 10))
np.random.seed(23)
# nx.draw_networkx_nodes(zachary, pos, node_color='cyan', node_size=350)
# nx.draw_networkx_edges(zachary, pos, edge_color='blue', alpha=0.5)
# nx.draw_networkx_labels(zachary, pos=pos, font_size=10, font_family="sans-serif")
nx.draw_networkx(zachary,with_labels=True,
                 node_size=500,
                 font_color="white", font_size=12,
                 node_color="darkblue", edgecolors="black",
                 font_weight="bold")
plt.axis("off")
plt.title("Zachary karate club")
plt.show()
```

#### Calculate degree centrality for all nodes of the graph

```python
# Calculate degree centrality
degree_centrality = nx.degree_centrality(zachary)

# Print the degree centrality measures for each node
print('Degree Centrality for all nodes:')
for node in list(zachary.nodes()):
    print("Node {}: {:0.4f}".format(node, degree_centrality[node]), end="   ,   ")
```

#### Plot the graph with node sizes based on degree centrality

```python
# Set plot area size
plt.figure(figsize=(12,8))

# Remove plot border
plt.axis('off')

# Set the random seed for layout
np.random.seed(23) # so every time we get the same layout
# Set the layout for the graph
pos = nx.kamada_kawai_layout(zachary)

# Determine node size based on degree centrality
degreecent_arr=np.array(list(degree_centrality.values()))

# Set the random seed for layout
np.random.seed(23)

# Set the title of the graph
plt.title("Zachary graph with node sizes based on degree centrality")


# Draw the graph
nx.draw_networkx(zachary,with_labels=True,
                 node_size=5000*degreecent_arr,
                 font_color="white", font_size=10,
                 node_color="darkblue", edgecolors="black",
                 font_weight="bold")
```

#### Calculate closeness centrality for all nodes of the graph

```python
# Calculate closeness centrality
closeness_centrality = nx.closeness_centrality(zachary)

# Print the closeness centrality measures for each node
print('Closeness Centrality for all nodes:')
for node in list(zachary.nodes()):
    print("Node {}: {:0.4f}".format(node, closeness_centrality[node]), end="   ,   ")
```

#### Plot the graph with node sizes based on closeness centrality

```python
# Set plot area size
plt.figure(figsize=(12,8))

# Remove plot border
plt.axis('off')

# Set the random seed for layout
#np.random.seed(23)
# Set the layout for the graph
#pos = nx.kamada_kawai_layout(zachary)

# Determine node size based on closeness centrality
closenesscent_arr=np.array(list(closeness_centrality.values()))


# Set the title of the graph
plt.title("Zachary graph with node sizes based on closeness centrality")

np.random.seed(23)
# Draw the graph
nx.draw_networkx(zachary,with_labels=True,
                 node_size=6000*(closenesscent_arr-0.26),
                 font_color="white", font_size=10,
                 node_color="darkblue", edgecolors="black",
                 font_weight="bold")
```

#### Calculate betweenness centrality for all nodes of the graph

```python
# Calculate betweenness centrality
betweenness_centrality = nx.betweenness_centrality(zachary)

# Print the betweenness centrality measures for each node
print('Betweenness Centrality for all nodes:')
for node in list(zachary.nodes()):
    print("node {}: {:0.4f}".format(node, betweenness_centrality[node]), end="   ,   ")
```

#### Plot the graph with node sizes based on betweenness centrality

```python
# Set plot area size
plt.figure(figsize=(12,8))

# Remove plot border
plt.axis('off')

# Set the layout for the graph
#pos = nx.kamada_kawai_layout(zachary)

# Determine node size based on betweenness centrality
betweennesscent_arr=np.array(list(betweenness_centrality.values()))


# Set the title of the graph
plt.title("Zachary graph with node sizes based on betweenness centrality")

np.random.seed(23)
# Draw the graph
nx.draw_networkx(zachary,with_labels=True,
                 node_size=5000*(betweennesscent_arr+0.04),
                 font_color="white", font_size=10,
                 node_color="darkblue", edgecolors="black",
                 font_weight="bold")
```

### Compare centrality measures for each node

```python
# Create a list of dictionaries containing the centrality measures for each node
data = []
for node in zachary.nodes():
    data.append({'Node': node, 'Degree Centrality': degree_centrality[node],
                 'Closeness Centrality': closeness_centrality[node],
                 'Betweenness Centrality': betweenness_centrality[node]})

# Create a pandas DataFrame from the list of dictionaries
df = pd.DataFrame(data)

# Print the DataFrame
print(df.to_string(index=False))
```

#### Calculate the correlation between Closeness and Betweenness centrality measures

```python
correlation = np.corrcoef(list(closeness_centrality.values()), list(betweenness_centrality.values()))[0, 1]
print(correlation)
```
